```python
# todo.py

# File to store tasks
TASKS_FILE = "tasks.txt"


# Load tasks from file
def load_tasks():
    try:
        with open(TASKS_FILE, "r") as f:
            tasks = [line.strip() for line in f.readlines()]
    except FileNotFoundError:
        tasks = []
    return tasks


# Save tasks to file
def save_tasks(tasks):
    with open(TASKS_FILE, "w") as f:
        for task in tasks:
            f.write(task + "\n")


# Add a task
def add_task(tasks):
    task = input("Enter a new task: ").strip()
    if task:
        tasks.append(task)
        save_tasks(tasks)
        print(f"‚úÖ Task added: {task}")
    else:
        print("‚ö†Ô∏è Task cannot be empty.")


# Remove a task
def remove_task(tasks):
    view_tasks(tasks)
    try:
        index = int(input("Enter task number to remove: ")) - 1
        if 0 <= index < len(tasks):
            removed = tasks.pop(index)
            save_tasks(tasks)
            print(f"üóëÔ∏è Task removed: {removed}")
        else:
            print("‚ö†Ô∏è Invalid task number.")
    except ValueError:
        print("‚ö†Ô∏è Enter a valid number.")


# View tasks
def view_tasks(tasks):
    if not tasks:
        print("üì≠ No tasks available.")
    else:
        print("\nYour To-Do List:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")
        print()


# Main menu
def main():
    tasks = load_tasks()

    while True:
        print("\n==== TO-DO LIST MENU ====")
        print("1. View tasks")
        print("2. Add task")
        print("3. Remove task")
        print("4. Exit")

        choice = input("Choose an option (1-4): ")

        if choice == "1":
            view_tasks(tasks)
        elif choice == "2":
            add_task(tasks)
        elif choice == "3":
            remove_task(tasks)
        elif choice == "4":
            print("üëã Exiting... Tasks saved.")
            break
        else:
            print("‚ö†Ô∏è Invalid choice, try again.")


if __name__ == "__main__":
    main()
```
